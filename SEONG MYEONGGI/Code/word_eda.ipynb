```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
# 데이터 읽기
train_df = pd.read_csv('/home/train.csv')
```

```python
# topic 단어수 별로 몇개의 데이터가 있는지 시각화(1~4개 제외)
train_df['topic_word_count'] = train_df['topic'].apply(lambda x: len(x.split()))
train_df_long_topics = train_df[train_df['topic_word_count'] >= 5]

plt.figure(figsize=(10, 6))
sns.countplot(x='topic_word_count', data=train_df_long_topics, palette='viridis')
plt.title('Number of Topics by Word Count')
plt.xlabel('Topic Word Count')
plt.ylabel('Number of Topics')
plt.show()
```
