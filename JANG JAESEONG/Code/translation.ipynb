```python
import pandas as pd
import os
```

```python
from deep_translator import GoogleTranslator

# 한국어에서 영어로 번역
translator = GoogleTranslator(source='ko', target='en')
translated_text = translator.translate("안녕하세요, 반갑습니다.")
print(translated_text)
```

```python
val_df = pd.read_csv(os.path.join('../data/','dev.csv'))
val_df.tail()
```

```python
train_df = pd.read_csv(os.path.join('../data/','train_eng_rework.csv'))
train_df.tail()
```

```python
#dev.csv 한영 번역 - dialogue
for n in range (499):
    a = val_df.loc[n,'dialogue']
    val_df.loc[n,'dialogue'] = translator.translate(a)
    print(n)
```

```python
# train.csv 한영 번역 - dialogue
for n in range (12457):
    a = train_df.loc[n,'dialogue']
    b = translator.translate(a)
    train_df.loc[n,'dialogue'] = b
    print(n)
```

```python
#train.csv 한영 번역 - topic
for n in range (12457):
    a = train_df.loc[n,'topic']
    b = translator.translate(a)
    train_df.loc[n,'topic'] = b
    print(b,n)
```

```python
#train.csv 한영 번역 - summary
for n in range (12457):
    a = train_df.loc[n,'summary']
    b = translator.translate(a)
    train_df.loc[n,'summary'] = b
    print(b,n)
```

```python
for n in range (499):
    a = val_df.loc[n,'summary']
    b = translator.translate(a)
    val_df.loc[n,'summary'] = b
    print(n)
```

```python
for n in range (499):
    a = val_df.loc[n,'topic']
    b = translator.translate(a)
    val_df.loc[n,'topic'] = b
    print(n)
```

```python
train_df.to_csv('train_eng.csv',encoding=False)
```

```python
val_df.to_csv('dev_eng.csv',encoding=False)
```
